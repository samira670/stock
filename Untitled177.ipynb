```python
import yfinance as yf

# Download stock data for a specific company (e.g., Apple, AAPL)
data = yf.download('AAPL', start='2010-01-01', end='2024-09-05')
print(data.head())
```

```python
# Select the closing price as a feature
data = data[['Close', 'Volume']]  # You can add other features too
```

```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler(feature_range=(0, 1))
scaled_data = scaler.fit_transform(data)
```

```python
import numpy as np

def create_sequences(data, time_steps):
    X, y = [], []
    for i in range(len(data) - time_steps):
        X.append(data[i:i + time_steps])
        y.append(data[i + time_steps, 0])  # Predict the closing price (or any target)
    return np.array(X), np.array(y)

time_steps = 100  # Use last 100 days as input for prediction
X, y = create_sequences(scaled_data, time_steps)
```

```python
X = np.reshape(X, (X.shape[0], X.shape[1], X.shape[2]))  # X is in (samples, time_steps, features)
```

```python
split = int(len(X) * 0.8)  # 80% for training, 20% for testing
X_train, X_test = X[:split], X[split:]
y_train, y_test = y[:split], y[split:]
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense, Dropout

# Build the LSTM model
model = Sequential()

# First LSTM layer with Dropout
model.add(LSTM(units=50, return_sequences=True, input_shape=(X_train.shape[1], X_train.shape[2])))
model.add(Dropout(0.2))

# Second LSTM layer
model.add(LSTM(units=50, return_sequences=False))
model.add(Dropout(0.2))

# Output layer
model.add(Dense(units=1))  # Predicting the next stock price

# Compile the model
model.compile(optimizer='adam', loss='mean_squared_error')

# Train the model
model.fit(X_train, y_train, epochs=50, batch_size=32)
```
